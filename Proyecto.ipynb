## Proyecto: Determinación del municipio con mejores ICFES en Ciencias Naturales



# Cargar de datos

```python
import pandas as pd
import matplotlib.pyplot as plt

archivo = '/content/Datos/Hoja1.xlsx'
df = pd.read_excel(archivo, skiprows=[0, 1, 2, 3, 4, 5], sheet_name='2023-4')
```

# Seleccionar datos

```python
df_subset = df.iloc[:,  [4, 5, 13] ]
df_subset.dropna( inplace=True )
df_subset =  df_subset.sort_values(  'PROMCIENCIASNATURALES', ascending=False  )
municipio = df_subset['NOMBREMUNICIPIO']
departamento = df_subset['DEPARTAMENTO']
promedio = df_subset['PROMCIENCIASNATURALES']
print(df_subset)
```

# Gráfica

```python
plt.plot(municipio[range(10)], promedio[range(10)])
plt.xlabel("Municipios")
plt.ylabel("Promedios")
plt.title("Municipio vs promedios")
plt.xticks(rotation=45)
plt.grid(linestyle='dotted', linewidth=2)
plt.show()
```
